```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('gari_import.csv')
df
```

```python
df = df.replace("Unknown", np.nan)
df = df.dropna(thresh=df.shape[1]-2)
df = df.reset_index(drop=True)
df
```

```python
df['Company'] = df['Company'].str.upper()
df
```

```python
df['Model'] = df['Model'].str.replace('-', '').str.capitalize()
df['Color'] = df['Color'].str.capitalize()
df['Fuel_Type'] = df['Fuel_Type'].str.capitalize()
df
```

```python
numeric_columns = ['Year', 'Engine_Capacity', 'Mileage', 'Price']  # Replace with your column names
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric).astype('Int64')
df.dtypes
```

```python
n_col = ['Year', 'Engine_Capacity']
n_df = df[n_col]
Q1 = n_df.quantile(0.25)
Q3 = n_df.quantile(0.75)
IQR = Q3 - Q1

df = df[~((n_df < (Q1 - 1.5 * IQR)) | (n_df > (Q3 + 1.5 * IQR))).any(axis=1)]
df
```

```python
df.to_csv('u_import')
```
